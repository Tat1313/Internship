In this dataset, we have 70 independent variables, so we need to perform the best set of feature selection for better interpretation and model performance. 

#### Univariate Selection

Statistical tests can be used to select those features that have the strongest relationship with output variable.

The Scikit-learn library provide the SlectKBest class that can be used with a suite of different statistical tests to select a specific number of features.

Chi-squared Statistical test

```python
from sklearn.feature_selection import SelectKBest
from sklearn.feature_selection import chi2
```

```python
x = dfCust_Retention.iloc[:,0:-1]
y = dfCust_Retention.iloc[:,-1]
```

```python
# apply SelectKBest to extract top 10 best Features
best_features = SelectKBest(score_func=chi2, k=10)
fit = best_features.fit(x,y)
```

```python
dfscores = pd.DataFrame(fit.scores_)
dfcolumns = pd.DataFrame(x.columns)
```

```python
# Concatenate two DataFrames for better visualization
dfBestFeatures = pd.concat([dfcolumns,dfscores],axis=1)
dfBestFeatures.columns = ['Best_Features',"Scores"]
```

```python
dfBestFeatures
```

```python
print(dfBestFeatures.nlargest(10,'Scores'))
```
